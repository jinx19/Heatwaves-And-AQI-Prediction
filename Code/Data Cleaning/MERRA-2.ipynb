```python
#Various Parameters from the MERRA-2 Dataset.
#10 different parameters are used for this dataset.
#All the parameters are same for the four cities.

parameters=Warangal_data['PARAMETER'].unique()
```

```python
#Displaying the 10 parameters.
#T2M: Temperature Monthly
#RH2M: Relative Humidity
#WS10M: Wind Speed at 10mts.
#T2M_MAX: Maximum Temperature.
#T2M_MIN: Minimum Temperature.
#WS10M_MAX: Maximum Windspeed at 10mts.
#WS10M_MIN: Minimum Windspeed at 10mts.
#WS50M_MAX: Maximum Windspeed at 50mts.
#WS50M_MIN: Minimum Windspeed at 50mts.
#P: Pressure.

parameters
```

```python
#Data Cleaning and Reformatting into proper rows.
Warangal=pd.DataFrame()
Nizamabad=pd.DataFrame()
Khammam=pd.DataFrame()
Karimnagar=pd.DataFrame()

for x in parameters:
    #Data is from 1981-2022.
    #There are 41 years.
    # 41 x 12 = 492 datapoints.
    
    #Warangal.
    mask = Warangal_data['PARAMETER'] == x
    # create a new DataFrame from the selected rows
    new_df =Warangal_data.loc[mask].copy()
    new_df.head(5)
    new_df = new_df.drop(['PARAMETER','YEAR'], axis=1)
    arr = np.array(new_df)
    # reshape the array
    new_arr = arr.reshape(492, 1)
    # convert the array back to DataFrame
    new_df = pd.DataFrame(new_arr, columns=[x])
    Warangal[x]=new_df
    
    #Nizamabad.
    mask = Nizamabad_data['PARAMETER'] == x
    # create a new DataFrame from the selected rows
    new_df =Nizamabad_data.loc[mask].copy()
    new_df.head(5)
    new_df = new_df.drop(['PARAMETER','YEAR'], axis=1)
    arr = np.array(new_df)
    # reshape the array
    new_arr = arr.reshape(492, 1)
    # convert the array back to DataFrame
    new_df = pd.DataFrame(new_arr, columns=[x])
    Nizamabad[x]=new_df
    
    #Khammam.
    mask = Khammam_data['PARAMETER'] == x
    # create a new DataFrame from the selected rows
    new_df =Khammam_data.loc[mask].copy()
    new_df.head(5)
    new_df = new_df.drop(['PARAMETER','YEAR'], axis=1)
    arr = np.array(new_df)
    # reshape the array
    new_arr = arr.reshape(492, 1)
    # convert the array back to DataFrame
    new_df = pd.DataFrame(new_arr, columns=[x])
    Khammam[x]=new_df
    
    #Karimnagar.
    mask = Karimnagar_data['PARAMETER'] == x
    # create a new DataFrame from the selected rows
    new_df =Karimnagar_data.loc[mask].copy()
    new_df.head(5)
    new_df = new_df.drop(['PARAMETER','YEAR'], axis=1)
    arr = np.array(new_df)
    # reshape the array
    new_arr = arr.reshape(492, 1)
    # convert the array back to DataFrame
    new_df = pd.DataFrame(new_arr, columns=[x])
    Karimnagar[x]=new_df
```
